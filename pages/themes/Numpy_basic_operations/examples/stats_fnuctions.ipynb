```python
import numpy as np
from matplotlib import pyplot as plt
```

## Percentile, Quontile

### Percentile
This function computes the q-th percentile of the data along the specified axis. The percentile is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations fall. For example, the 50th percentile is the median.

### Quantile

This function computes the q-th quantile of the data along the specified axis. Quantiles are essentially the same as percentiles but are expressed as fractions or decimals rather than percentages. For example, the 0.5 quantile is equivalent to the 50th percentile.

```python
data = [1, 2, 2, 3, 4, 4, 4, 5]

print(f'data len: {len(data)}')
print(np.percentile(data, 50))
print(np.quantile(data, 0.5))
```

```python
# Given data
hours = [1, 2, 2, 3, 4, 5, 5, 6, 7, 7, 7, 7, 7, 9]
print(f'hours len: {len(hours)}')

# Percentiles and quantiles to plot
percentiles = np.percentile(hours, [25, 50, 75])
quantiles = np.quantile(hours, [0.25, 0.5, 0.75])

print(percentiles)
print(quantiles)

# Plotting the data
plt.figure(figsize=(12, 6))

# Original data
plt.plot(hours, label='Hours', marker='o')

# Percentiles
for percentile in percentiles:
    plt.axhline(y=percentile, linestyle='--', color='r', label=f'Percentile {percentile}')

# Quantiles
for quantile in quantiles:
    plt.axhline(y=quantile, linestyle='-.', color='b', label=f'Quantile {quantile}')

plt.legend()
plt.title('Percentiles and Quantiles of Hours Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
